```python
# First let's import the packages we will use in this project
# You can do this all now or as you need them
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)  # Adjust the configuration of the plots

pd.options.mode.chained_assignment = None


# Now we need to read in the data
df = pd.read_csv(r"C:\Users\user\Downloads\movies.csv")
```

```python
# Now let's take a look at the data

df.head()
```

```python
# We need to see if we have any missing data
# Let's loop through the data and see if there is anything missing

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
df = df.dropna()
```

```python
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
# Data Types for our columns

print(df.dtypes)
```

```python
# Change datatype of the columns

df['budget'] = df['budget'].astype('int64')

df['gross'] = df['gross'].astype('int64')
```

```python
df.head()
```

```python
# Create correct year column

# Uses a regular expression to extract only the year from the date

df['yearcorrect'] = df['released'].str.extract(pat = '([0-9]{4})').astype(int)
df.head()
```

```python
# Order our Data a little bit to see

df = df.sort_values(by=['gross'], inplace=False, ascending=False)
df.head()
```

```python
# To look at all of the data

pd.set_option('display.max_rows', None)
```

```python
# Drop any duplicates

df.drop_duplicates().head()
```

```python
# Scatter plot with budget vs gross

plt.figure(figsize=(8, 5))

plt.scatter(x=df['budget'], y=df['gross'], s=10)

plt.title('Budget vs Gross Earnings')

plt.xlabel('Budget for film')

plt.ylabel('Gross Earnings')

plt.show()
```

```python
df.head()
```

```python
# Plot budget vs gross using seaborn

plt.figure(figsize=(8, 5))

sns.regplot(x='budget', y='gross', data=df, scatter_kws={'s': 10}, line_kws={'color': 'blue'})

plt.show()
```

```python
# Correlation Matrix between all numeric columns

df.corr(numeric_only=True, method ='pearson')
```

```python
correlation_matrix = df.corr(numeric_only=True, method ='pearson')

plt.figure(figsize=(8, 5))

sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix for Numeric Features")

plt.xlabel("Movie features")

plt.ylabel("Movie features")

plt.show()
```

```python
df_numerized = df.copy()

for col_name in df_numerized.select_dtypes(include=['object']).columns:
    df_numerized[col_name] = df_numerized[col_name].astype('category').cat.codes
    

df_numerized.head()
```

```python
df.head()
```

```python
correlation_matrix = df_numerized.corr(numeric_only=True, method ='pearson')

plt.figure(figsize=(14, 8))

sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix for Numeric Features")

plt.xlabel("Movie features")

plt.ylabel("Movie features")

plt.show()
```

```python
df_numerized.corr()
```

```python
# Unstacking

correlation_mat = df_numerized.corr()

corr_pairs = correlation_mat.unstack()

corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()

sorted_pairs
```

```python
high_corr = sorted_pairs[(sorted_pairs) > 0.5]

high_corr
```
